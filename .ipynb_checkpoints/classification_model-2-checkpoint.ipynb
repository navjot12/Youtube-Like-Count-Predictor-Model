```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
ds = pd.read_csv('../new_data.csv')
# ds = ds.drop(ds[ds.viewCount >10000].index)
ds.head()
```

```python
del ds['videoId']
del ds['channelId']
del ds['channelTitle']
del ds['title']
del ds['publishedAt']
del ds['currentTime']
del ds['caption']
del ds['allowed']
del ds['recordingDate']
del ds['allowedCount']
del ds['publicStatsViewable']
del ds['favoriteCount']
del ds['privacyStatus']
del ds['dimension']
del ds['latitude']
del ds['longitude']
del ds['definition']
del ds['durationCategory']
del ds['license']
del ds['licensedContent']
del ds['embeddable']
del ds['description']
del ds['categoryId']
ds.head()
```

```python
ds = ds.dropna(subset=['likeCount', 'viewCount'])

ds['life'] = ds['life'].astype('uint')
ds['life'] = ds['life'].apply(lambda x: x/(60*60*24))

ds.head()
```

```python
mean_Comm = ds['commentCount'].mean()
ds['commentCount'] = ds['commentCount'].fillna(int(mean_Comm))
ds.describe()
```

```python
data = np.asarray([ds['life'], ds['duration'], ds['commentCount'], ds['viewCount'], ds['dislikeCount'], ds['likeCount']]).T
data = (data - data.mean(axis=0))/data.std(axis=0)
print data.shape
corr = np.cov(data.T).astype('float')[-1]
print corr
```

```python
records = np.asarray([ds['life'], ds['duration'], ds['commentCount'], ds['viewCount'], ds['dislikeCount']]).T
records = (records - records.mean(axis=0))/records.std(axis=0)
ans = np.asarray(ds['likeCount']).T

split = int(0.9*records.shape[0])
X_train = records[:split, :]
y_train = ans[:split]

X_test = records[split:, :]
y_test = ans[split:]

print X_train.shape, y_train.shape
print X_test.shape, y_test.shape
```

```python
def dist(x1, x2):
    return np.sqrt(((x1-x2)**2).sum())           #divide this distance by range to normalize things?
    # return np.sqrt(0.5*((np.sqrt(x1) - np.sqrt(x2))**2).sum())
    

def knn(X_train, x, y_train, k=5):
    vals = []
    for ix in range(X_train.shape[0]):
        v = [dist(x, X_train[ix, :]), y_train[ix]]
        vals.append(v)
    
    updated_vals = sorted(vals, key=lambda x:x[0])
    pred_arr = np.asarray(updated_vals[:k])
    pred_arr = pred_arr[1]
    return pred_arr.mean(axis=0)

def whknn(x_train, x, y_train, k=5):
    vals = []
    for ix in range(x_train.shape[0]):
        di = dist(x, x_train[ix, :])
        wi = 1/(di + (10**(-6)))
        v = [wi, y_train[ix]]
        vals.append(v)

    updated_vals = sorted(vals, key=lambda x:x[0], reverse=True)
    pred_arr = np.asarray(updated_vals[:k])
    total_weight = pred_arr.sum(axis=0)[0]
    prod = 0
    for ix in range(pred_arr.shape[0]):
        prod += pred_arr[ix, 0]*pred_arr[ix, 1]
    prod /=total_weight
    return prod
```

```python
std = []
y_pred = []
for ix in range(X_test.shape[0]):
    q = X_test[ix]
    res = knn(X_train, q, y_train, k=4)
    y_pred.append(res)
    std.append(np.abs(res - y_test[ix]))
    # print ix, std[ix]

plt.figure(0)
plt.plot(std)
plt.show()
print np.mean(std)
```

```python
from sklearn.metrics import r2_score
print r2_score(y_test, y_pred)
```
